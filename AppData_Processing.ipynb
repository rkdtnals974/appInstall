```python
## Kaggle에서 다운받은 csv파일을 read_csv()로 읽어온다
import pandas as pd


app_data = pd.read_csv("googleplaystore.csv")
app_data
```

```python
## Rating의 NaN값을 지운다

app_data1=app_data.dropna(subset=["Rating"],axis=0)
app_data1
```

```python
## Category column 확인 중 "1.9"라는 동떨어진 값이 있어서 해당 row 확인
pd.unique(app_data1["Category"])
```

```python
## 데이터가 밀려있는 row로 간주하고 삭제 진행
print(app_data1[app_data1["Category"] == "1.9"])
```

```python
app_data2 = app_data1.drop(app_data1[app_data1["Category"] == "1.9"].index)
app_data2
```

```python
## 데이터의 정보 확인
app_data2.info()
```

```python
## 데이터분석을 위한 Reviews, Price, Size의 타입을 바꿔준다

## Reviews: object -> float
app_data2['Reviews'] =  pd.to_numeric(app_data2['Reviews'])

## Price: object -> float
price = app_data2['Price'].replace(r'\$','', regex=True)
app_data2['Price'] = pd.to_numeric(price)

## Size: object -> float
## 단위가 Mb와 Kb로 나뉘어져 있기 때문에 단위를 Kb로 통일하고 "k"를 제거한다.
import numpy as np
size = app_data2["Size"]
regex_expression = r"\d+(?:.\d+)?"
units_size = size.replace(regex_expression, '', regex=True)

units_size = units_size.replace({'M':1000, 'k': 1, 'Varies with device': np.nan})
app_data2["Size"] = pd.to_numeric(app_data2["Size"].replace(r"M|k", "", regex=True), errors='coerce').multiply(units_size)
```

```python
## Installs의 값 확인

pd.unique(app_data2["Installs"])
```

```python
## Installs의 값이 너무 많기에 범위를 지정함

app_data2.loc[app_data2["Installs"] == "500+", "Installs"] = "1 ~ 1,000"
app_data2.loc[app_data2["Installs"] == "1+", "Installs"] = "1 ~ 1,000"
app_data2.loc[app_data2["Installs"] == "5+", "Installs"] = "1 ~ 1,000"
app_data2.loc[app_data2["Installs"] == "10+", "Installs"] = "1 ~ 1,000"
app_data2.loc[app_data2["Installs"] == "50+", "Installs"] = "1 ~ 1,000"
app_data2.loc[app_data2["Installs"] == "100+", "Installs"] = "1 ~ 1,000"

app_data2.loc[app_data2["Installs"] == "1,000+", "Installs"] = "1,000 ~ 10,000"
app_data2.loc[app_data2["Installs"] == "5,000+", "Installs"] = "1,000 ~ 10,000"

app_data2.loc[app_data2["Installs"] == "10,000+", "Installs"] = "10,000 ~ 100,000"
app_data2.loc[app_data2["Installs"] == "50,000+", "Installs"] = "10,000 ~ 100,000"

app_data2.loc[app_data2["Installs"] == "100,000+", "Installs"] = "100,000 ~ 1,000,000"
app_data2.loc[app_data2["Installs"] == "500,000+", "Installs"] = "100,000 ~ 1,000,000"

app_data2.loc[app_data2["Installs"] == "1,000,000+", "Installs"] = "1,000,000 ~ 10,000,000"
app_data2.loc[app_data2["Installs"] == "5,000,000+", "Installs"] = "1,000,000 ~ 10,000,000"

app_data2.loc[app_data2["Installs"] == "10,000,000+", "Installs"] = "10,000,000 ~ 100,000,000"
app_data2.loc[app_data2["Installs"] == "50,000,000+", "Installs"] = "10,000,000 ~ 100,000,000"

app_data2.loc[app_data2["Installs"] == "100,000,000+", "Installs"] = "100,000,000 ~"
app_data2.loc[app_data2["Installs"] == "500,000,000+", "Installs"] = "100,000,000 ~"
```
